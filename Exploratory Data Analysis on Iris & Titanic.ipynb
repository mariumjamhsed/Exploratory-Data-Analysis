Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
```

Importing datasets

```python
titanic=sns.load_dataset("titanic")
titanic
```

```python
iris=sns.load_dataset("iris")
iris
```

```python
tips=sns.load_dataset("tips")
tips
```

Visualizations

## kde plot on the basis of total bill and gender

```python
sns.displot(data=tips,x="total_bill",kind="kde",fill=True,hue="sex",height=6,aspect=2)
```

## Scatter plot on the basis of p-class and age

```python
sns.relplot(data=titanic,x="pclass",y="age",height=6,aspect=2,kind="scatter",hue="adult_male")
```

Scatter plot on the basis of the tips and total bill further categorised in gender and smoker

```python
sns.relplot(data=tips,x="total_bill",y="tip",row="sex",col="smoker")
```

```python
sns.histplot(data=iris,bins=20)
```

```python
sns.histplot(data=titanic,x="survived",bins=20)
```

## strip plot on titanic dataset

```python
sns.catplot(data=titanic,x="fare",y="class",kind="strip",jitter=False)
```

## bivariate KDE plot

```python
sns.kdeplot(data=titanic,x="fare",y="survived")
```

```python
gap=px.data.gapminder()
```

```python
gap
```

```python
asia=gap[gap["continent"]=="Asia"]
```

```python
asia
```

```python
asia=asia.pivot(index="country",columns="year",values="lifeExp")
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(asia)
```

## Heatmap(Europe)

```python
temp_df=gap[gap["continent"]=="Europe"].pivot(index="country", columns="year", values="lifeExp")
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(temp_df,alpha=0.5,linewidths=0.2,annot=True,cmap="YlGnBu")
```
